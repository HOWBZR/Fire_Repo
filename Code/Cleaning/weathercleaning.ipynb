```python
import pandas as pd
```

```python
weather_df = pd.read_csv('../Data/WeatherEvents_Jan2016-Dec2021.csv')
```

```python
weather_df.head()
```

```python
weather_df['Type'].value_counts()
```

```python
weather_df['EventId'].value_counts()
```

```python
weather_df.drop(columns = ['EventId'], inplace =True)
```

```python
weather_df.head()
```

```python
weather_df.isna().sum()
```

```python
weather_df['Year'] = weather_df['StartTime(UTC)'].str.split('-', expand = True)[0]
```

```python
weather_df['Month'] = weather_df['StartTime(UTC)'].str.split('-', expand = True)[1]
```

```python
weather_df['Day'] = weather_df['StartTime(UTC)'].str.split('-', expand = True)[2]
```

```python
weather_df.head()
```

```python
weather_df['Day'] = weather_df['Day'].str.split(' ', expand = True)[0]
```

```python
weather_df.head()
```

```python
weather_df.drop(columns = ['StartTime(UTC)','EndTime(UTC)'], inplace =True)
```

```python
weather_df.drop(columns = ['TimeZone','AirportCode'], inplace =True)
```

```python
weather_df.info()
```

```python
weather_df.drop(columns = ['City','ZipCode'], inplace =True)
```

```python
weather_df.head()
```

```python
weather_df['State'].value_counts()
```

```python
weather_df.head()
```

```python
weather_df.to_csv('../Data/weather_cleaned.csv') 
```
